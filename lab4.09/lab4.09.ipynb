```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

sns.set_style("ticks",{'axes.grid' : True})
# sns.set_theme(rc={'figure.figsize':(5, 4)})
```

### Задание 1

```python
df = pd.read_csv("task1.csv", sep=";")
df["alpha_rad"] = np.deg2rad(df["alpha"])
df["cos"] = np.cos(df["alpha_rad"])
df["cos**2"] = df["cos"] ** 2
df["I (мкА)"] /= 100
```

```python
df["cos"] = round(df["cos"], 2)
df["cos**2"] = round(df["cos**2"], 2)
df["I (мкА)"] /= 100
print(df[["alpha", "cos", "cos**2", "I (мкА)"]].to_latex(index=False, float_format="%.4f"))
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(np.cos(df["alpha_rad"]) ** 2, df["I (мкА)"], zorder=2.0, label=r"$(cos^2(\alpha), I)$")
plt.plot(np.cos(df["alpha_rad"]) ** 2, df["I (мкА)"], zorder=2.0)
plt.grid(zorder=1.0)
plt.xlabel(r"$x=cos^2(\alpha)$")
plt.ylabel(r"$I$, " + "мкА")
plt.xticks(np.arange(0, 1.1, 0.1))
# plt.yticks(np.arange(0, max(df["I (мкА)"]) + 1.5, 1.5))
plt.legend()
plt.show()
```
